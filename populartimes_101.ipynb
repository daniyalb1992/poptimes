```python
import pandas as pd 
import json
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import csv file (getting errors with other files - csv format issue ....)
data = pd.read_csv("C:\\Users\\Daniyal\\Desktop\\Popular Times\\poptimes\\3.csv", sep =";", encoding="ISO-8859-1")
```

```python
# separate hourly nestted data into columns of the dataframe 
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
for i in range(0,24):
    data["monday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][0]['data'][i])

for i in range(0,24):
    data["tuesday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][1]['data'][i])

for i in range(0,24):
    data["wednesday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][2]['data'][i])

for i in range(0,24):
    data["thursday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][3]['data'][i])

for i in range(0,24):
    data["friday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][4]['data'][i])

for i in range(0,24):
    data["saturday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][5]['data'][i])

for i in range(0,24):
    data["sunday{0}".format(i)] = data['popular_times'].transform(lambda x: json.loads(x)['populartimes'][6]['data'][i])
    
# loops for creating group columns     
for day in days:
    data["furher_{0}".format(day)] =  data["{}{}".format(day,0)] + data["{}{}".format(day,1)] + data["{}{}".format(day,2)] + data["{}{}".format(day,3)]
    
for day in days:
    data["morgens_{0}".format(day)] =  data["{}{}".format(day,4)] + data["{}{}".format(day,5)] + data["{}{}".format(day,6)] + data["{}{}".format(day,7)]
    
for day in days:
    data["mittag_{0}".format(day)] =  data["{}{}".format(day,8)] + data["{}{}".format(day,9)] + data["{}{}".format(day,10)] + data["{}{}".format(day,11)]
    
for day in days:
    data["nachmittags_{0}".format(day)] =  data["{}{}".format(day,12)] + data["{}{}".format(day,13)] + data["{}{}".format(day,14)] + data["{}{}".format(day,15)]
    
for day in days:
    data["abends_{0}".format(day)] =  data["{}{}".format(day,16)] + data["{}{}".format(day,17)] + data["{}{}".format(day,18)] + data["{}{}".format(day,19)]
    
for day in days:
    data["mitternacht_{0}".format(day)] =  data["{}{}".format(day,20)] + data["{}{}".format(day,21)] + data["{}{}".format(day,22)] + data["{}{}".format(day,23)]
```

```python
# separate dataframes containing group information for each day ... 
monday = data[['id', 'title', 'furher_monday', 'morgens_monday', 'mittag_monday', 'nachmittags_monday', 'abends_monday', 'mitternacht_monday']].copy()
tuesday = data[['id', 'title', 'furher_tuesday', 'morgens_tuesday', 'mittag_tuesday', 'nachmittags_tuesday', 'abends_tuesday', 'mitternacht_tuesday']].copy()
wednesday = data[['id', 'title', 'furher_wednesday', 'morgens_wednesday', 'mittag_wednesday', 'nachmittags_wednesday', 'abends_wednesday', 'mitternacht_wednesday']].copy()
thursday = data[['id', 'title', 'furher_thursday', 'morgens_thursday', 'mittag_thursday', 'nachmittags_thursday', 'abends_thursday', 'mitternacht_thursday']].copy()
friday = data[['id', 'title', 'furher_friday', 'morgens_friday', 'mittag_friday', 'nachmittags_friday', 'abends_friday', 'mitternacht_friday']].copy()
saturday = data[['id', 'title', 'furher_saturday', 'morgens_saturday', 'mittag_saturday', 'nachmittags_saturday', 'abends_saturday', 'mitternacht_saturday']].copy()
sunday = data[['id', 'title', 'furher_sunday', 'morgens_sunday', 'mittag_sunday', 'nachmittags_sunday', 'abends_sunday', 'mitternacht_sunday']].copy()
```

```python
# calculate standard deviations across each store for each day's dataframe ... 
monday['std.dev'] = monday.iloc[:,:].std(1)
tuesday['std.dev'] = tuesday.iloc[:,:].std(1)
wednesday['std.dev'] = wednesday.iloc[:,:].std(1)
thursday['std.dev'] = thursday.iloc[:,:].std(1)
friday['std.dev'] = friday.iloc[:,:].std(1)
saturday['std.dev'] = saturday.iloc[:,:].std(1)
sunday['std.dev'] = sunday.iloc[:,:].std(1)
```
